```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler
```

```python
#####VERİ SETİNİN İNDİRİLMESİ##########
otomobil=pd.read_csv(r'C:\Users\Cenk Toker\OneDrive\Desktop\yapay_zeka_yüksek_lisans\ikinci dönem\istatistiksel yapay zeka\final\otomobil.csv')
```

```python
otomobil.describe()
```

```python
#####BAĞIMLI DEĞİŞKEN##########
y=otomobil["SatisFiyati"]
```

```python
#####BAĞIMLI DEĞİŞKEN##########
X=otomobil.drop(["SatisFiyati"], axis=1)
```

```python
#####VERİYİ STANDARTLAŞTIRMA##########
sutunlar=X.columns
X=MinMaxScaler().fit_transform(X)
X=pd.DataFrame(X , columns=sutunlar)
X.describe()
```

```python
########## LİNEER MODELİN OLUŞTURULMASI 
X=sm.add_constant(X)
reg_2=sm.OLS(y,X).fit()
```

```python
reg_2.summary()
```

```python
###### BU SONUÇLARA BAKILDIĞINDA ,ANLAMLILIK DÜZEYİNİ %10 SEÇTİĞİMİZİ DÜŞÜNÜRSEK,  F İSTATİSTİĞİNİN P VALUE'SU 0.0879 
###### ANLAMLIDIR. BU BAKIŞ AÇISI İLE VERİLERİN EN AZ BİRİSİNİN SATIŞ FİYATINI AÇIKLADIĞI DÜŞÜNÜLEBİLİR
###### ANCAK , ADJUSTED R SQUARE OLDUKÇA DÜŞÜK. BAĞIMLI DEĞİŞKENDEKİ VARYASYONUN %3,1 İ AÇIKLANIYOR GİBİ GÖRÜNÜYOR. 
###### BAĞIMSIZ DEĞİŞKENLERDEN İSE , P VALUE'LARI %10 DAN KÜÇÜK OLAN GARANTİ SÜRESİ , KAPI SAYISI , MODEL YILI , VE SABİT DEĞER 
###### ANLAMLI GORUNUYOR. BU DEĞİŞKENLERLE TEKRAR BİR REGRESYON DENEYECEĞİM.
```

```python
#####ANLAMSIZ BAĞIMSIZ DEĞİŞKENLERİN MODELDEN UZAKLAŞTIRILMASI##########
X=X.drop(["YakıtTüketimi" , "MotorGucu" , "Kilometre" , "KoltukSayisi" , "MaxHiz" , "AracKodu"], axis=1)
```

```python
########## LİNEER MODELİN OLUŞTURULMASI 
reg_3=sm.OLS(y,X).fit()
```
