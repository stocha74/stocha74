```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',1000)
```

```python
#####VERİ SETİNİN İNDİRİLMESİ##########
konut=pd.read_csv(r'C:\Users\Cenk Toker\OneDrive\Desktop\yapay_zeka_yüksek_lisans\ikinci dönem\istatistiksel yapay zeka\final\konut.csv')
```

```python
konut.describe()
```

```python
#####BAĞIMLI DEĞİŞKEN##########
y=konut["Fiyat"]
```

```python
#####BAĞIMLI DEĞİŞKEN##########
X=konut.drop(["Fiyat" , "Port_no"], axis=1)
```

```python
X
```

```python
#####VERİYİ STANDARTLAŞTIRMA##########
sutunlar=X.columns
```

```python
X=MinMaxScaler().fit_transform(X)
```

```python
X=pd.DataFrame(X , columns=sutunlar)
```

```python
X.describe()
```

```python
#########VALİDASYONSUZ MODELİN OLUŞTURULMASI , TAHMİNLEME VE MAE HESAPLANMASI ######
model=LinearRegression()
model.fit(X,y)
pred=model.predict(X)
mae_validasyonsuz=mae(y,pred)
model1=pd.DataFrame({"Kaysayı_model1":model.coef_ }, index=sutunlar)
model1.loc["Hata_Model1"]=mae_validasyonsuz
```

```python
#####EĞİTİM VE TEST VERİLEİNİ AYIRMA##############
###### OKUL NO : 20221500658 İÇİN ################
####### TEST ORANI DIGITLER TOPLAMI %31 OLACAK ###
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.31)
```

```python
#########VALİDASYONLU MODELİN OLUŞTURULMASI , TAHMİNLEME VE MAE HESAPLANMASI ######
model=LinearRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
mae_validasyonlu=mae(y_test,pred)
model2=pd.DataFrame({"Kaysayı_model2":model.coef_ }, index=sutunlar)
model2.loc["Hata_Model2"]=mae_validasyonlu
```

```python
model_son=pd.concat([model1,model2],axis=1)
```
