```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
#####VERİ SETİNİN İNDİRİLMESİ##########
dataset=pd.read_csv(r'C:\Users\Cenk Toker\OneDrive\Desktop\yapay_zeka_yüksek_lisans\ikinci dönem\istatistiksel yapay zeka\final\classification_dataset.csv')
```

```python
##### DATASET'E VE İSTATİSTİKSEL DEĞERLERİNE BAKILDIĞINDA
##### BAĞIMLI DEĞİŞKEN (LABEL) AYRILDIKTAN SONRA , 
##### NORMALİZE ETMEYE GEREK KALMAYACAĞI GÖRÜLMEKTEDİR
```

```python
#####BAĞIMLI DEĞİŞKEN##########
y=dataset["Label"]
```

```python
#####BAĞIMLI DEĞİŞKEN##########
X=dataset.drop(["Label"], axis=1)
```

```python
X
```

```python
####FONKSİYON
def egitim_test_orani_belirleme(soyisim):
    import random
    deger=len(soyisim)
    random.seed(deger)
    egitim_test_orani=round(random.uniform(0.1 , 0.8),2)
    return egitim_test_orani
```

```python
#### EĞİTİM VE TEST GRUPLARININ FONKSİYONUN GÖSTERDİĞİ ORANDA BÖLÜNMESİ
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=egitim_test_orani_belirleme("toker"))
```

```python
##### Support Vector Machine kullanılarak tahmin yapılması ve accuracy değerinin elde edilmesi
svm=SVC()
svm.fit(X_train,y_train)
svm_pred=svm.predict(X_test)
svm_acc=accuracy_score(y_test,svm_pred)
```

```python
##### Random Forest kullanılarak tahmin yapılması ve accuracy değerinin elde edilmesi
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
rf_pred=rf.predict(X_test)
rf_acc=accuracy_score(y_test,rf_pred)
```

```python
##### Gradient Boosting kullanılarak tahmin yapılması ve accuracy değerinin elde edilmesi
gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
gb_pred=gb.predict(X_test)
gb_acc=accuracy_score(y_test,gb_pred)
```

```python
##### Decision Tree kullanılarak tahmin yapılması ve accuracy değerinin elde edilmesi
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt_pred=dt.predict(X_test)
dt_acc=accuracy_score(y_test,dt_pred)
```

```python
####SONUÇLARIN TOPLU OLARAK DATAFRAME'E YAZILMASI
sonuc={
        "SVC" : svm_acc,
        "RF":rf_acc,
        "GB":gb_acc,
        "DT":dt_acc
}
```

```python
sonuc_df=pd.DataFrame(list(sonuc.items()),columns=["Model","Accuracy"])
```

```python
sonuc_df=sonuc_df.sort_values(by="Accuracy" , ascending=False)
```
